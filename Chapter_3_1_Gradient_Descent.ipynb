```python
# Deep Learning 101 for Scientists and Engineers
# Chapter 3.1: Gradient Descent
# Yong-Jun Shin, 2024

import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = x^2 - 2*x + 1
def f(x):
    return x**2 - 2*x + 1

# Derivative of the function f(x)
def df(x):
    return 2*x - 2

# Gradient descent parameters
x_start = 0  # starting point
learning_rate = 0.1  # step size
num_steps = 10  # number of iterations

# Arrays to store the steps
x_values = [x_start]
f_values = [f(x_start)]

# Perform gradient descent
x = x_start
for i in range(num_steps):
    grad = df(x)  # Get the derivative
    x -= learning_rate * grad  # Update x
    x_values.append(x)
    f_values.append(f(x))

# Plotting the function
x_range = np.linspace(0, 1, 100)
plt.plot(x_range, f(x_range), label='f(x) = x^2 - 2*x + 1')

# Plotting the gradient descent path
plt.scatter(x_values, f_values, color='blue')
plt.plot(x_values, f_values, color='blue', label='Path by gradient descent')

# Adding titles and labels
plt.title('Gradient Descent Optimization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
