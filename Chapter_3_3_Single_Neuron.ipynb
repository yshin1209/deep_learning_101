```python
# Deep Learning 101 for Scientists and Engineers
# Chapter 3.3: Single Nueron
# Yong-Jun Shin, 2024

import torch

# Define the data and parameter values
x = torch.tensor(4.0)  
y = torch.tensor(2.0)
a = torch.tensor(0.0, requires_grad=True)

# Estimated (predicted) output y_est
y_est = a * x

# Estimation (prediction) error e
e = y - y_est

# Cost function: e^2 
cost = e ** 2
print(f"Cost: {cost.item()}")
```

```python
# Perform backpropagation
cost.backward()
print(f"Gradient of a: {a.grad.item()}")
```

```python
# Learning rate for the optimizer
learning_rate = 0.1

# Define the optimizer
optimizer = torch.optim.SGD([a], lr=learning_rate)

# Update parameters using SGD
optimizer.step()
print(f"Updated value of a: {a.item()}")
```

```python
import torch
import matplotlib.pyplot as plt

# Define the data and parameter values
x = torch.tensor(4.0)
y = torch.tensor(2.0)
a = torch.tensor(0.0, requires_grad=True)

# Learning rate for the optimizer
learning_rate = 0.01

# Define the optimizer
optimizer = torch.optim.SGD([a], lr=learning_rate)

# Lists to store the values of 'a' and cost for plotting
a_values = []
cost_values = []

# Loop to update the parameter 'a' 100 times
for i in range(100):
    # Estimated (predicted) output y_est
    y_est = a * x

    # Estimation (prediction) error e
    e = y - y_est

    # Cost function: e^2
    cost = e ** 2

    # Perform backpropagation
    optimizer.zero_grad()  # Reset gradients to zero to avoid accumulation
    cost.backward()

    # Update parameters using SGD
    optimizer.step()

    # Append current value of 'a' and cost
    a_values.append(a.item())
    cost_values.append(cost.item())

# Plotting the results
plt.figure(figsize=(10, 5))

# Plotting the evolution of 'a'
plt.subplot(1, 2, 1)
plt.plot(a_values)
plt.title('Value of a over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Value of a')

# Plotting the cost over iterations
plt.subplot(1, 2, 2)
plt.plot(cost_values)
plt.title('Cost over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Cost')

plt.tight_layout()
plt.show()
```
